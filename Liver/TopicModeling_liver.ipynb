```r
Sys.setenv(LANG="en")
library(Matrix)
library(fastTopics)
library(ggplot2)
library(cowplot)
set.seed(1)
```

```r
datapath<-'input_vizgen_liver/alldata/'
datapath
```

```r
counts_sc <- read.csv(paste(datapath,'sc_liver_data_downsample.csv',sep=""), sep=",", header=TRUE,row.names=1)
dim(counts_sc)
```

```r
class(counts_sc)
```

```r
counts_sp <- read.csv(paste(datapath,'spatial_quadrant.csv',sep=""),sep=",", header=TRUE, row.names=1)
dim(counts_sp)
```

```r
class(counts_sp)
```

```r
counts_sp <- Matrix(as.matrix(counts_sp))
counts_sc <- Matrix(as.matrix(counts_sc))
```

```r
fit <- fit_topic_model(counts_sp,k = 6)
```

```r
names(fit)
```

```r
apply(fit$F,2,summary)
```

```r
fit_sc <- fit_topic_model(counts_sc,k = 6)
```

```r
topic_colors <- c("skyblue","forestgreen","darkmagenta","dodgerblue",
                  "gold","darkorange")
structure_plot(fit_sc,colors = topic_colors)
```

```r
#spatial cells 
structure_plot(fit,colors = topic_colors)
```

```r
#single cell 
set.seed(1)
pca <- prcomp(fit_sc$L)$x
clusters <- kmeans(pca,centers = 6,iter.max = 100)$cluster
summary(factor(clusters))

structure_plot(fit_sc,topics = 1:6,colors = topic_colors,gap = 25,
               grouping = clusters)
```

```r
#spatial 
set.seed(1)
pca <- prcomp(fit$L)$x
clusters <- kmeans(pca,centers = 6,iter.max = 100)$cluster
summary(factor(clusters))

structure_plot(fit,topics = 1:6,colors = topic_colors,gap = 25,
               grouping = clusters)
```
